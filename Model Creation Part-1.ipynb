```python

import pandas as pd
import numpy as np
import seaborn as sns
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
dataset.bmi.replace(to_replace=np.nan, value=dataset.bmi.mean(), inplace=True)
```

```python
dataset.describe()
```

```python
dataset= dataset.drop(columns=['id'])
#dataset.info()
X = dataset.drop(columns=['stroke'])  # Input features
Y = dataset['stroke']
X
```

```python
from sklearn.preprocessing import OrdinalEncoder
X['gender'].unique()
X
```

```python
gender = ['Female','Male','Other']
enc = OrdinalEncoder(categories=[gender])
X[['gender']] = enc.fit_transform(X[['gender']])
X
```

```python
X['ever_married'].unique()
ever_married = ['No','Yes']
enc = OrdinalEncoder(categories=[ever_married])
X[['ever_married']] = enc.fit_transform(X[['ever_married']])
X
```

```python
X['work_type'].unique()
work_type = ['children','Govt_job','Self-employed','Private','Never_worked']
enc = OrdinalEncoder(categories=[work_type])
X[['work_type']] = enc.fit_transform(X[['work_type']])
X



```

```python
X['Residence_type'].unique()
Residence_type = ['Urban','Rural']
enc = OrdinalEncoder(categories=[Residence_type])
X[['Residence_type']] = enc.fit_transform(X[['Residence_type']])
X
```

```python
X['smoking_status'].unique()
smoking_status = ['never smoked','formerly smoked','smokes', 'Unknown']
enc = OrdinalEncoder(categories=[smoking_status])
X[['smoking_status']] = enc.fit_transform(X[['smoking_status']])
X



```

```python
x=X.to_numpy()
y=Y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 0)
```

```python
print("Number transactions x_train dataset: ", x_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions x_test dataset: ", x_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
x_train_res, y_train_res = sm.fit_resample(x_train, y_train.ravel())

print('After OverSampling, the shape of train_X: {}'.format(x_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_res = scaler.fit_transform(x_train_res)
x_test= scaler.transform(x_test)
```

# Model creating

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, classification_report, roc_curve, auc, precision_recall_curve, average_precision_score
from sklearn.model_selection import cross_val_score
```

```python
##model selection
from sklearn import metrics
from sklearn.metrics import recall_score, f1_score, precision_score
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(12, input_shape=(10,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python

# Fit the model to the training data
model.fit(x_train_res, y_train_res, epochs=50, batch_size=10)

y_pred = model.predict(x_test)
# Threshold predictions
y_pred_binary = (y_pred > 0.5).astype(int)


# Calculate other metrics
recall = recall_score(y_test, y_pred_binary, average='weighted')
f1 = f1_score(y_test, y_pred_binary, average='weighted')
precision = precision_score(y_test, y_pred_binary, average='weighted')

# Print the metrics
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
print(f'Precision: {precision:.2f}')
```

```python
new_person_data = [[ 1.0,67.0,0.0,1.0,1.0 ,3.0 ,0.0,200,36,1.0 ]]
prediction = model.predict(new_person_data)
rounded = [round(x[0]) for x in prediction]
print(rounded)
```

```python
import tensorflow as tf
from tensorflow import keras, lite

# Corrected quotation marks
keras_file = "Stroke_Prediction.h5"
tf.keras.models.save_model(model, keras_file)

converter = lite.TFLiteConverter.from_keras_model(model)
tfmodel = converter.convert()

open("Stroke_Prediction.tflite", "wb").write(tfmodel)
```
